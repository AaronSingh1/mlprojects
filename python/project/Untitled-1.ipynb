```python
import tkinter as tk
from PIL import Image, ImageDraw
import numpy as np
from tensorflow.keras.models import load_model

# Load the trained model
model = load_model('C:/Users/aaron/OneDrive/Pictures/MNIST_keras_CNN.h5')  # Replace with your actual model file

# Create the GUI window
window = tk.Tk()
window.title("Handwritten Digit Recognition")
window.geometry("400x400")

# Create the canvas for drawing
canvas = tk.Canvas(window, width=280, height=280, bg="white")
canvas.pack()

# Create an empty PIL image and a PIL draw object
image = Image.new("L", (280, 280), color=255)
draw = ImageDraw.Draw(image)

# Function to handle mouse movements
def on_mouse_move(event):
    x, y = event.x, event.y
    r = 8  # Radius of the drawn point
    draw.ellipse((x - r, y - r, x + r, y + r), fill=0)
    canvas.create_oval(x - r, y - r, x + r, y + r, fill="black")

# Function to handle digit recognition
def recognize_digit():
    resized_image = image.resize((28, 28))
    grayscale_image = resized_image.convert("L")
    inverted_image = Image.eval(grayscale_image, lambda x: 255 - x)
    input_data = np.array(inverted_image).reshape((1, 28, 28, 1)).astype('float32') / 255.0
    prediction = np.argmax(model.predict(input_data))
    result_label.config(text=f"Predicted digit: {prediction}")

# Bind mouse movement event to the canvas
canvas.bind("<B1-Motion>", on_mouse_move)

# Create a button to trigger digit recognition
recognize_button = tk.Button(window, text="Recognize Digit", command=recognize_digit)
recognize_button.pack()

# Create a label to display the prediction result
result_label = tk.Label(window, text="")
result_label.pack()

# Start the GUI event loop
window.mainloop()
```
